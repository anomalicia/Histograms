```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import exp
import random

random.seed(3)
```

Evaluate whether the descriptive statistics provided useful information about the variable. Can you identify any common characteristics of the distributions that could be usefully described using the mean and/or standard deviation, versus the ones that could not? 

All of the distributions chosen, except for Wald and Pareto, follow a bell-shaped curve with a visually identfiable center. The mean and standard deviation are not as useful for the Wald and Pareto distributions because they aren't normalized.

```python
%matplotlib inline
h = np.random.wald(3, 2, 100000)
plt.hist(np.random.wald(3, 2, 100000), bins=200, density=True)
plt.axvline(h.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(h.mean()+h.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(h.mean()-h.std(), color='b', linestyle='dashed', linewidth=2)
plt.title('Wald or Inverse Gaussian Distribution')
plt.xlabel('Random Values')
plt.show()
```

```python
%matplotlib inline
n, p = 10, .5  # number of trials, probability of each trial
s = np.random.binomial(n, p, 1000)
plt.hist(np.random.binomial(n, p, 1000), bins=20, density=True)
plt.axvline(s.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(s.mean()+s.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(s.mean()-s.std(), color='b', linestyle='dashed', linewidth=2)
plt.title('Binomial Distribution')
plt.xlabel('Random Values')
plt.show()
```

```python
%matplotlib inline
mu, kappa = 0.0, 4.0 # mean and dispersion
s = np.random.vonmises(mu, kappa, 1000)
plt.hist(np.random.vonmises(mu, kappa, 1000), bins=20, density=True)
plt.axvline(s.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(s.mean()+s.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(s.mean()-s.std(), color='b', linestyle='dashed', linewidth=2)
plt.title('Von Mises Distribution')
plt.xlabel('Random Values')
plt.show()
```

```python
%matplotlib inline
a, m = 3., 2.  # shape and mode
s = (np.random.pareto(a, 1000) + 1) * m

plt.hist(((np.random.pareto(a, 1000) + 1) * m), bins=20, density=True)
plt.axvline(s.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(s.mean()+s.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(s.mean()-s.std(), color='b', linestyle='dashed', linewidth=2)
plt.title('Pareto Distribution')
plt.xlabel('Random Values')
plt.show()
```

```python
s = np.random.poisson(5, 10000)

plt.hist(np.random.poisson(5, 10000), bins=20, density=True)
plt.axvline(s.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(s.mean()+s.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(s.mean()-s.std(), color='b', linestyle='dashed', linewidth=2)
plt.title('Poisson Distribution')
plt.xlabel('Random Values')
plt.show()
```

```python
s = np.random.rayleigh(3, 100000)
plt.hist(np.random.rayleigh(3, 100000), bins=200, density=True)
plt.axvline(s.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(s.mean()+s.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(s.mean()-s.std(), color='b', linestyle='dashed', linewidth=2)
plt.title('Rayleigh Distribution')
plt.xlabel('Random Values')
plt.show()
```

Evaluate the descriptive statistics against the data. 

It makes sense that the sum of the two normalized histograms would be in the center, with a higher frequency at the center than either of the two histograms alone. The mean and standard deviation are also centered.

```python
x = np.random.normal(5, .5, 1000)
y = np.random.normal(10, 1, 1000)
z = (x + y) / 2

plt.hist(x, color='blue', bins=np.arange(2, 16), alpha=.5) 
plt.hist(y, color='red', bins=np.arange(2, 16), alpha=.5)
plt.hist(z, color='purple', bins=np.arange(2, 16), alpha=.5)
plt.axvline(z.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(z.mean()+z.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(z.mean()-z.std(), color='b', linestyle='dashed', linewidth=2)
plt.title('Normed histograms')
plt.xlabel('Random Values')

plt.show()
# Seed because random.seed
```
