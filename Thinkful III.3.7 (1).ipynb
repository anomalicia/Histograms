```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
%matplotlib inline
```

```python
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 

plt.hist(pop1, alpha=0.5, label='Population 1', color='blue', edgecolor='black', linewidth=1) 
plt.hist(pop2, alpha=0.5, label='Population 2', color='red', edgecolor='black', linewidth=1) 
plt.legend(loc='upper right')
plt.title('Population size: 10000')
plt.show()
```

```python
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

plt.hist(sample1, alpha=0.5, label='Sample 1', color='blue', edgecolor='black', linewidth=1) 
plt.hist(sample2, alpha=0.5, label='Sample 2', color='red', edgecolor='black', linewidth=1) 
plt.legend(loc='upper right') 
plt.title('Sample size: 100')
plt.show()
```

```python
print('The sample1 mean is:', sample1.mean())
print('The sample2 mean is:', sample2.mean())
print('The sample1 standard deviation is:', sample1.std())
print('The sample2 standard deviation is:', sample2.std())

# Compute the difference between the two sample means.
diff=sample2.mean( ) -sample1.mean()
print('The difference between the two sample means is: ', diff)
```

```python
size = np.array([len(sample1), len(sample2)])
sd = np.array([sample1.std(), sample2.std()])

# The squared standard deviations are divided by the sample size and summed, then we take
# the square root of the sum. 
diff_se = (sum(sd ** 2 / size)) ** 0.5  

#The difference between the means divided by the standard error: T-value.  
print('The T-value is:', diff/diff_se)
```

```python
from scipy.stats import ttest_ind
print(ttest_ind(sample2, sample1, equal_var=False))
```

Change the probability value (p) for pop1 to 0.3, then take new samples and compute the t-statistic and p-value. Then change the probability value p for group 1 to 0.4, and do it again. What changes, and why? 

It seems that as the pop1 p increases, or gets closer to the pop2 p, the values become more aligned and p-value increases. The t-value decreases as samples become more similar.

```python
pop1 = np.random.binomial(10, 0.3, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(pop1, alpha=0.5, label='Population 1', color='blue', edgecolor='black', linewidth=1) 
plt.hist(pop2, alpha=0.5, label='Population 2', color='red', edgecolor='black', linewidth=1) 
plt.legend(loc='upper right') 

sample7 = np.random.choice(pop1, 100, replace=True)
sample8 = np.random.choice(pop2, 100, replace=True)

plt.subplot(1, 2, 2)
plt.hist(sample7, alpha=0.5, label='Sample 7', color='blue', edgecolor='black', linewidth=1) 
plt.hist(sample8, alpha=0.5, label='Sample 8', color='red', edgecolor='black', linewidth=1) 
plt.legend(loc='upper right') 
plt.show()
```

```python
pop1 = np.random.binomial(10, 0.4, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 

# Let’s make histograms for the two groups. 

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(pop1, alpha=0.5, label='Population 1', color='blue', edgecolor='black', linewidth=1) 
plt.hist(pop2, alpha=0.5, label='Population 2', color='red', edgecolor='black', linewidth=1) 
plt.legend(loc='upper right') 

sample9 = np.random.choice(pop1, 100, replace=True)
sample10 = np.random.choice(pop2, 100, replace=True)

plt.subplot(1, 2, 2)
plt.hist(sample9, alpha=0.5, label='Sample 9', color='blue', edgecolor='black', linewidth=1) 
plt.hist(sample10, alpha=0.5, label='Sample 10', color='red', edgecolor='black', linewidth=1) 
plt.legend(loc='upper right') 
plt.show()
```

```python
print(ttest_ind(sample8, sample7, equal_var=False))
diff3 = sample8.mean() - sample7.mean()
size = np.array([len(sample7), len(sample8)])
sd = np.array([sample7.std(), sample8.std()])
diff_se3 = (sum(sd ** 2 / size)) ** 0.5  
print('The T-value is:', diff3/diff_se3)
```

```python
print(ttest_ind(sample10, sample9, equal_var=False))
diff4 = sample10.mean() - sample9.mean()
size = np.array([len(sample9), len(sample10)])
sd = np.array([sample9.std(), sample10.std()])
diff_se4 = (sum(sd ** 2 / size)) ** 0.5  
print('The T-value is:', diff4/diff_se4)
```
